## NLP ASSIGNMENT 1
### AIM : To apply Bag of Words, TF-IDF, and N-gram techniques on a UCI dataset for text feature extraction.

Step 1 : import all the lib and load the dataset

https://www.kaggle.com/datasets/saurabhshahane/ecommerce-text-classification

```python
import pandas as pd

# Use a raw string (r"...") to avoid issues with backslashes and special characters
df = pd.read_csv(r"C:\Users\kndak\Downloads\archive (1)\ecommerceDataset.csv")

# Preview the first 5 rows
print(df.head())
```

```python
# Keep only the text column (assuming it's the second column)
df = df.iloc[:, 1]

# Rename the column for clarity
df = df.to_frame(name="text")

# Show the cleaned dataframe
print(df.head())
```

```python
df.dropna(inplace=True)
df = df[df["text"].str.strip() != ""]
```

### Step 2: Preprocess Text - Basic text cleaning (lowercase, remove punctuation/numbers)

```python
import re
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# Basic text cleaning (lowercase, remove punctuation/numbers)
df["cleaned_text"] = df["text"].apply(lambda x: re.sub(r"[^a-zA-Z\s]", "", x.lower()))
```

```python
print(df.iloc[0])
```

### Step 3: Bag of Words (BoW)

```python
# Initialize CountVectorizer for BoW
bow_vectorizer = CountVectorizer(max_features=1000)  # Limit features for performance
bow_matrix = bow_vectorizer.fit_transform(df["cleaned_text"])
```

```python
# View BoW as DataFrame
bow_df = pd.DataFrame(bow_matrix.toarray(), columns=bow_vectorizer.get_feature_names_out())
print(bow_df.head())
```

```python
# Initialize TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=1000)
tfidf_matrix = tfidf_vectorizer.fit_transform(df["cleaned_text"])

# View TF-IDF as DataFrame
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
print(tfidf_df.head())
```

```python
# Bigrams using CountVectorizer
bigram_vectorizer = CountVectorizer(ngram_range=(2, 2), max_features=500)
bigram_matrix = bigram_vectorizer.fit_transform(df["cleaned_text"])

# View bigrams
bigram_df = pd.DataFrame(bigram_matrix.toarray(), columns=bigram_vectorizer.get_feature_names_out())
print(bigram_df.head())
```
